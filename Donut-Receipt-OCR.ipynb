```python
"""
CORD Receipt -> Structured Data Pipeline (v4, Donut, Google Colab)
====================================================================
Uses naver-clova-ix/donut-base-finetuned-cord-v2 -- an OCR-free
vision-encoder/decoder model already fine-tuned specifically on the
CORD dataset. It reads receipt pixels directly and generates structured
JSON (menu items, subtotal, total) -- no OCR engine, no regex parsing,
no preprocessing pipeline needed. This is the highest-accuracy option
for this exact dataset short of training your own model.

Colab setup (run in the first cell)
------------------------------------
!pip install transformers accelerate sentencepiece rapidfuzz -q

Runtime -> Change runtime type -> GPU  (T4 is enough)

Getting the dataset into Colab (pick ONE):

Option A -- Kaggle API:
    from google.colab import files
    files.upload()  # upload your kaggle.json
    !mkdir -p ~/.kaggle && cp kaggle.json ~/.kaggle/ && chmod 600 ~/.kaggle/kaggle.json
    !kaggle datasets download -d aqqqrfs/cord-dataset -p /content
    !unzip -q /content/cord-dataset.zip -d /content/cord-dataset

Option B -- Google Drive:
    from google.colab import drive
    drive.mount('/content/drive')
    # then set DATASET below to your Drive path

Then set DATASET to wherever the CORD/ folder (with train/dev/test
subfolders, each containing image/ and json/) ended up.

IMPORTANT accuracy caveat: this checkpoint was fine-tuned using CORD's
own train split, so scores on `train` will look artificially perfect.
Trust the `dev`/`test` numbers as the real accuracy signal.
"""

import re
import json
import time
import pandas as pd
import torch
from pathlib import Path
from PIL import Image
from rapidfuzz import fuzz
from tqdm import tqdm
from transformers import DonutProcessor, VisionEncoderDecoderModel

# ---------------------------------------------------------------------------
# 1. Dataset paths -- EDIT THIS to wherever you put the CORD folder in Colab
# ---------------------------------------------------------------------------
DATASET = Path("/content/cord_dataset/CORD")


def load_split(split):
    image_dir = DATASET / split / "image"
    json_dir = DATASET / split / "json"

    images = sorted(image_dir.glob("*.png"))
    jsons = sorted(json_dir.glob("*.json"))

    json_map = {j.stem: j for j in jsons}

    matched_images = []
    matched_jsons = []
    missing_json = []
    for img in images:
        if img.stem in json_map:
            matched_images.append(img)
            matched_jsons.append(json_map[img.stem])
        else:
            missing_json.append(img.name)

    image_stems = {img.stem for img in images}
    missing_image = [j.name for j in jsons if j.stem not in image_stems]

    print(f"{split}: images={len(images)} jsons={len(jsons)} matched={len(matched_images)}")
    if missing_json:
        print(f"  Missing JSON ({len(missing_json)}):", missing_json[:5])
    if missing_image:
        print(f"  Missing images ({len(missing_image)}):", missing_image[:5])

    return matched_images, matched_jsons


train_images, train_jsons = load_split("train")
dev_images, dev_jsons = load_split("dev")
test_images, test_jsons = load_split("test")

print("\nFinal dataset sizes:")
print("Train:", len(train_images))
print("Dev:", len(dev_images))
print("Test:", len(test_images))

# ---------------------------------------------------------------------------
# QUICK SANITY CHECK MODE -- verify everything works before a full run
# ---------------------------------------------------------------------------
SAMPLE_SIZE = None  # e.g. SAMPLE_SIZE = 5 for a quick smoke test first

if SAMPLE_SIZE is not None:
    train_images, train_jsons = train_images[:SAMPLE_SIZE], train_jsons[:SAMPLE_SIZE]
    dev_images, dev_jsons = dev_images[:SAMPLE_SIZE], dev_jsons[:SAMPLE_SIZE]
    test_images, test_jsons = test_images[:SAMPLE_SIZE], test_jsons[:SAMPLE_SIZE]
    print(f"\n[SAMPLE MODE] Using only {SAMPLE_SIZE} images per split.")


# ---------------------------------------------------------------------------
# 2. Donut model (loaded once)
# ---------------------------------------------------------------------------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"\n[Donut] device = {DEVICE}"
      + ("" if DEVICE == "cuda" else "  WARNING: no GPU detected, this will be slow. "
                                     "Runtime -> Change runtime type -> GPU."))

MODEL_NAME = "naver-clova-ix/donut-base-finetuned-cord-v2"
processor = DonutProcessor.from_pretrained(MODEL_NAME)
model = VisionEncoderDecoderModel.from_pretrained(MODEL_NAME)
model.to(DEVICE)
model.eval()

TASK_PROMPT = "<s_cord-v2>"


@torch.no_grad()
def run_donut(image_path: Path) -> dict:
    """Run Donut on one receipt image, return its parsed structured dict
    (schema: {'menu': [...], 'sub_total': {...}, 'total': {...}})."""
    image = Image.open(image_path).convert("RGB")
    pixel_values = processor(image, return_tensors="pt").pixel_values.to(DEVICE)

    decoder_input_ids = processor.tokenizer(
        TASK_PROMPT, add_special_tokens=False, return_tensors="pt"
    ).input_ids.to(DEVICE)

    outputs = model.generate(
        pixel_values,
        decoder_input_ids=decoder_input_ids,
        max_length=model.decoder.config.max_position_embeddings,
        early_stopping=True,
        pad_token_id=processor.tokenizer.pad_token_id,
        eos_token_id=processor.tokenizer.eos_token_id,
        use_cache=True,
        num_beams=1,
        bad_words_ids=[[processor.tokenizer.unk_token_id]],
        return_dict_in_generate=True,
    )

    sequence = processor.batch_decode(outputs.sequences)[0]
    sequence = sequence.replace(processor.tokenizer.eos_token, "")
    sequence = sequence.replace(processor.tokenizer.pad_token, "")
    sequence = re.sub(r"<.*?>", "", sequence, count=1).strip()  # drop task token
    try:
        return processor.token2json(sequence)
    except Exception:
        return {}


# ---------------------------------------------------------------------------
# 3. Normalize Donut's output schema into {total, items} for comparison
# ---------------------------------------------------------------------------
def normalize_amount(s):
    if s is None:
        return None
    s = re.sub(r"[^\d.,]", "", str(s)).replace(",", "").strip()
    if not s:
        return None
    try:
        return float(s)
    except ValueError:
        return None


def donut_to_common(parsed: dict) -> dict:
    items = []
    menu = parsed.get("menu", [])
    if isinstance(menu, dict):
        menu = [menu]
    if isinstance(menu, list):
        for m in menu:
            if isinstance(m, dict):
                nm = m.get("nm")
                price = normalize_amount(m.get("price"))
                if nm:
                    items.append({"name": nm, "price": price})

    total_price = None
    total_field = parsed.get("total", {})
    if isinstance(total_field, list):
        total_field = total_field[0] if total_field else {}
    if isinstance(total_field, dict):
        total_price = normalize_amount(total_field.get("total_price"))

    return {"total": total_price, "items": items}


# ---------------------------------------------------------------------------
# 4. Ground truth (same CORD valid_line schema as before)
# ---------------------------------------------------------------------------
def _entry_text(entry, values_only=True):
    words = entry.get("words", [])
    texts = []
    for w in words:
        if values_only and w.get("is_key", 0) == 1:
            continue
        t = w.get("text", "")
        if t:
            texts.append(t)
    return " ".join(texts).strip()


def load_ground_truth(json_path: Path):
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    valid_lines = data.get("valid_line", [])
    groups = {}
    for entry in valid_lines:
        gid = entry.get("group_id")
        groups.setdefault(gid, []).append(entry)

    items = []
    total_price = None
    for gid, entries in groups.items():
        cat_map = {}
        for e in entries:
            cat_map[e.get("category", "")] = _entry_text(e)

        if "menu.nm" in cat_map or "menu.price" in cat_map:
            nm = cat_map.get("menu.nm")
            price = normalize_amount(cat_map.get("menu.price"))
            if nm:
                items.append({"name": nm, "price": price})

        if "total.total_price" in cat_map:
            total_price = normalize_amount(cat_map["total.total_price"])

    return {"total": total_price, "items": items}


# ---------------------------------------------------------------------------
# 5. Build structured dataset
# ---------------------------------------------------------------------------
def build_dataset(images, jsons, split_name):
    records = []
    times = []
    for img_path, json_path in tqdm(list(zip(images, jsons)), desc=f"Donut [{split_name}]", unit="img"):
        t0 = time.time()
        raw_pred = run_donut(img_path)
        times.append(time.time() - t0)

        pred = donut_to_common(raw_pred)
        gt = load_ground_truth(json_path)

        records.append(
            {
                "image_id": img_path.stem,
                "split": split_name,
                "pred_total": pred["total"],
                "pred_items": json.dumps(pred["items"]),
                "gt_total": gt["total"],
                "gt_items": json.dumps(gt["items"]),
            }
        )
    if times:
        print(f"[{split_name}] avg {sum(times) / len(times):.2f}s/image over {len(times)} images")
    return pd.DataFrame(records)


# ---------------------------------------------------------------------------
# 6. Evaluation (same metric logic as the OCR versions, for direct comparison)
# ---------------------------------------------------------------------------
def amounts_match(a, b, tol=0.01):
    if a is None or b is None:
        return False
    return abs(a - b) <= tol


def items_prf(pred_items, gt_items, price_tol=0.01, name_thresh=70):
    if not gt_items and not pred_items:
        return 1.0, 1.0, 1.0
    matched_gt = set()
    tp = 0
    for p in pred_items:
        for i, g in enumerate(gt_items):
            if i in matched_gt:
                continue
            price_ok = amounts_match(p["price"], g["price"], price_tol)
            name_ok = fuzz.token_set_ratio(str(p["name"]), str(g["name"])) >= name_thresh
            if price_ok and name_ok:
                matched_gt.add(i)
                tp += 1
                break
    precision = tp / len(pred_items) if pred_items else 0.0
    recall = tp / len(gt_items) if gt_items else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0.0
    return precision, recall, f1


def evaluate(df):
    total_hits, total_total = 0, 0
    precisions, recalls, f1s = [], [], []

    for _, row in df.iterrows():
        gt_total = row["gt_total"]
        if gt_total is not None:
            total_total += 1
            if amounts_match(row["pred_total"], gt_total):
                total_hits += 1

        pred_items = json.loads(row["pred_items"])
        gt_items = json.loads(row["gt_items"])
        p, r, f1 = items_prf(pred_items, gt_items)
        precisions.append(p)
        recalls.append(r)
        f1s.append(f1)

    return {
        "total_accuracy": total_hits / total_total if total_total else None,
        "item_precision_avg": sum(precisions) / len(precisions) if precisions else None,
        "item_recall_avg": sum(recalls) / len(recalls) if recalls else None,
        "item_f1_avg": sum(f1s) / len(f1s) if f1s else None,
        "n_samples": len(df),
    }


# ---------------------------------------------------------------------------
# 7. Run
# ---------------------------------------------------------------------------
print("\n--- Running Donut structured extraction ---")
train_df = build_dataset(train_images, train_jsons, "train")
dev_df = build_dataset(dev_images, dev_jsons, "dev")
test_df = build_dataset(test_images, test_jsons, "test")

full_df = pd.concat([train_df, dev_df, test_df], ignore_index=True)
full_df.to_csv("/content/cord_donut_structured_output.csv", index=False)
print("Saved:", full_df.shape, "-> /content/cord_donut_structured_output.csv")

print("\n=== Evaluation: Train (NOTE: model was fine-tuned on train -- expect inflated scores) ===")
print(evaluate(train_df))
print("\n=== Evaluation: Dev (trust this) ===")
print(evaluate(dev_df))
print("\n=== Evaluation: Test (trust this) ===")
print(evaluate(test_df))

eval_summary = pd.DataFrame(
    [
        {"split": "train", **evaluate(train_df)},
        {"split": "dev", **evaluate(dev_df)},
        {"split": "test", **evaluate(test_df)},
    ]
)
eval_summary.to_csv("/content/cord_donut_eval_summary.csv", index=False)
print("\nSaved -> /content/cord_donut_eval_summary.csv")
```
